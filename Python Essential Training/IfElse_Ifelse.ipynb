# If and Else

### If statements with "FizzBuzz"

```python
# 1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, FizzBuzz, 16

for n in range (1, 101):
    if n % 5 == 0:
        print('FizzBuzz')
    else:
        if n % 3 == 0:
            print('Fizz')
        else:
            if n % 5 == 0:
                print('Buzz')
            else:
                print(n)
        
```

### Elif statements

```python
for n in range (1, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 5 == 0:
        print('Buzz')
    elif n % 3 == 0:
        print('Fizz')
    else:
        print(n)
```

### Single Line if statements

```python
# one liner if else but can be kind of confusing 
n = 5
print('Fizz' if n % 3 == 0 else n)
```

```python
# can also stick in a variable
# this is called "Ternary Operator"
FizzBuzz = 'Fizz' if n % 3 == 0 else n
```
